# The Bernoulli distribution

We will now apply the coin toss experiment to a broader context.
Consider a test with two possible results: $0$ and $1$.
Consider $0$ to be "failure" and $1$ to be "success".

Assume for the moment that the experiment has a probability of $\theta$, which is a value between $0$ and $1$.
Let us take a fair coin $\theta=0.5$ as an example.
However, any value is acceptable here.
The following random variable captures the outcome of such an experiment:


$$
X = \begin{cases}
1,\;\text{with probability}\;\theta,\\
0,\;\text{otherwise}.
\end{cases}
$$

In terms of the probability mass function, we have:

$$
p(X=0) = \theta,
$$

and

$$
p(X=1) = 1 - \theta.
$$

This random variable has a special name.
It is called a Bernoulli random variable, named after [Jacob Bernoulli](https://en.wikipedia.org/wiki/Jacob_Bernoulli).
Another say that $X$ is a Bernoulli random variable is to write:

$$
X\sim \text{Bernoulli}(\theta).
$$

which is read as:

> The random variable $X$ *follows* a Bernoulli *distribution* with parameter
> $\theta$.

This is enough information to figure out what the probability mass function is.
For example, with a Google search "Bernoulli distribution" you can get
[here](https://en.wikipedia.org/wiki/Bernoulli_distribution).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st

# Matplotlib and Seaborn settings for better plot quality
%matplotlib inline
sns.set(rc={"figure.dpi":100, 'savefig.dpi':300})
sns.set_context('notebook')
sns.set_style("ticks")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina', 'svg')
```

I will pick $\theta=0.6$:

```python
theta = 0.6
X = st.bernoulli(theta)
```

This objects encapsulates everything related to the random variable $X$.
For example, `X.support()` tells you which values it takes:

```python
X.support()
```

To evaluate the probability mass function, then we can use the following:

```python
X.pmf(0)
```

```python
X.pmf(1)
```

sample random data :

```python
z=X.rvs(100)
```

Let's now take many samples and do their histogram:

```python
xs = X.rvs(size=100)
fig, ax = plt.subplots(dpi=300)
ax.hist(xs)
ax.set_xlabel('$x$')
ax.set_ylabel('Counts');
ax.hist(xs,color='green', edgecolor='black')
```
