***this a histogram showing carbon emissions as they are extracted in the data set  in co2***

```python
co2_data=df['co2']
print(co2_data)
sns.set_style('whitegrid')
sns.histplot(data=co2_data,bins=20,color='gray')
plt.xlabel ('carbon dioxide Emission')
plt.ylabel('Frequency')
plt.title('Distribution of carbon ')
plt.show()
```

**This histogram shows clearly how each fuel  type contributes to carbon emissions**

```python
fuel= df['fuelType']
sns.histplot(data=df, x='fuelType', color='purple',bins=20, edgecolor='black')
plt.xlabel('Fuel type  Efficiency')
plt.ylabel('Frequency')
plt.title('Contribution of types of fuel ')
plt.savefig('Contribution of types of fuel')
plt.show()
print(fuel.head(10))
```

**this histogram clearly shows the cities with high rates of emmission gases**

```python
city= df['city']
sns.histplot(data=df, x='city', color='navy',bins=20, edgecolor='blue')
plt.xlabel('Cities rates')
plt.ylabel('Frequency')
plt.title('Contribution of Cities ')
plt.savefig('Contribution of Cities')
plt.show()
print(city.head(10))
```

**this code snipet shows the history of all columns creating correlation heatmap for each of them**

```python
figures= df.select_dtypes(include=['number'])
corr_matrix = figures.corr()
plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Fuel Economy Data')
plt.savefig('emission diff')
plt.show()
```
